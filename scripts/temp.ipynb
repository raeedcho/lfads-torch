```python
import h5py


data_path = '../datasets/brian_marmoset_data.h5'
# data_path = '../runs/lfads-torch-example/nlb_mc_maze/250328_exampleSingle/lfads_output_mc_maze-20ms-val.h5'
with h5py.File(data_path, "r") as h5file:
    data_dict = {k: v[()] for k, v in h5file.items()}

data_dict.keys()
```

```python
import numpy as np
corr = np.corrcoef(np.row_stack(list(data_dict['train_encod_data'])),rowvar=False)
import seaborn as sns
sns.heatmap(corr, square=True, cmap='coolwarm', cbar_kws={"shrink": .8})
```
